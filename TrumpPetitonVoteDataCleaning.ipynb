```python
import pandas as pd 
import json 
```

```python
with open('brexit_petition.json') as data_file:
    brexit_votes_data = json.load(data_file)
    
with open('votes.json') as data_file:    
    votes_data = json.load(data_file)

with open('populations.json') as data_file:    
    population_data = json.load(data_file)
```

```python
votes = pd.DataFrame(votes_data['data']['attributes']['signatures_by_constituency']).set_index('ons_code')
brexit_votes = pd.DataFrame(brexit_votes_data['data']['attributes']['signatures_by_constituency']).set_index('ons_code')
```

```python
populations = pd.DataFrame({"ons_code": list(population_data.keys()), "pop" : [v['population'] for v in population_data.values()],"constituency" :[v['constituency'] for v in population_data.values()]}).set_index('ons_code')
```

```python
votes.join(populations).join(brexit_votes["signature_count"], rsuffix='_brexit').to_csv('trump_petition_signatures.csv')
```
